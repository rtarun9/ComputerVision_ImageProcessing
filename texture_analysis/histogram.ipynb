```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from pathlib import Path
import os
```

```python
frames = []

for child in Path('../data/stereo_video/').iterdir():
    if child.is_file():
        frames.append(cv2.cvtColor(cv2.imread('../data/stereo_video/' + child.name), cv2.COLOR_BGR2RGB))
plt.imshow(frames[0])
```

```python
def plot_histogram(image):
    # plt.hist(image.ravel(), bins=256, range=(0, 255), fc='k', ec='k')
    plt.show()
    print(image.max())
    vals = [0] * 255
    for i in range(image.shape[0]-1):
        for j in range(image.shape[1]-1):
            if image[i][j] == 0:
                continue
            vals[image[i][j]] = vals[image[i][j]] + 1

    # plot histogram with 255 bins
    plt.plot(vals)
    plt.show()
```

```python
plot_histogram(cv2.cvtColor(frames[0], cv2.COLOR_RGB2GRAY))
```

```python
images = []

for child in Path('../data/segmented_images/').iterdir():
    if child.is_file():
        images.append(cv2.imread('../data/segmented_images/' + child.name))
        images[-1] = cv2.cvtColor(images[-1], cv2.COLOR_RGB2GRAY)
    
        print(child.name)
        plt.imshow(images[-1], cmap='gray')
        plt.show()

        print('histogram')
        plot_histogram(images[-1])
```
