```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

def draw_sift_matches(image, template):
    img1 = template         
    img2 = image 
    # Initiate SIFT detector
    sift = cv.SIFT_create()
    # find the keypoints and descriptors with SIFT
    kp1, des1 = sift.detectAndCompute(img1,None)
    kp2, des2 = sift.detectAndCompute(img2,None)
    # BFMatcher with default params
    bf = cv.BFMatcher()
    matches = bf.knnMatch(des1,des2,k=2)
    # Apply ratio test
    good = []
    for m,n in matches:
        if m.distance < 0.75*n.distance:
            good.append([m])
    # cv.drawMatchesKnn expects list of lists as matches.
    img3 = cv.drawMatchesKnn(img1,kp1,img2,kp2,good,None,flags=cv.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
    plt.imshow(img3),plt.show()
```

```python
stop_images_pass_cases = []
stop_images_pass_cases.append(cv.imread('../data/Dataset-SIFT/Stop/pass1.png', 0))
stop_images_pass_cases.append(cv.imread('../data/Dataset-SIFT/Stop/pass2.png', 0))
stop_images_pass_cases.append(cv.imread('../data/Dataset-SIFT/Stop/pass3.png', 0))
stop_images_pass_cases.append(cv.imread('../data/Dataset-SIFT/Stop/pass4.png', 0))
stop_images_pass_cases.append(cv.imread('../data/Dataset-SIFT/Stop/pass5.png', 0))
stop_images_pass_cases.append(cv.imread('../data/Dataset-SIFT/Stop/pass6.png', 0))

stop_image_fail_cases = []
stop_image_fail_cases.append(cv.imread('../data/Dataset-SIFT/Stop/fail1.png', 0))
stop_image_fail_cases.append(cv.imread('../data/Dataset-SIFT/Stop/fail2.png', 0))
stop_image_fail_cases.append(cv.imread('../data/Dataset-SIFT/Stop/fail3.png', 0))
stop_image_fail_cases.append(cv.imread('../data/Dataset-SIFT/Stop/fail4.png', 0))

stop_image_template = cv.imread('../data/Dataset-SIFT/Stop/template.png', 0)

print("PASS CASES")
for image in stop_images_pass_cases:
    draw_sift_matches(image, stop_image_template)
    
print("\n\n\n\n\nFAIL CASES")
for image in stop_image_fail_cases:
    draw_sift_matches(image, stop_image_template)
    
```
