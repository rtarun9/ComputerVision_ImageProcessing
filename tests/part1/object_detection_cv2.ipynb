```python
# CB.EN.U4CSE20666
# Tarun Ramaswamy
# Changes done : Fixed the smile detection (currently, detected_smile is a empty tuple)
# -*- coding: utf-8 -*-

"""
Created on Fri Apr 21 09:07:53 2023
HAAR CASCADE CLASSIFIERS
Object Detection with OpenCV-Python Using a Haar-Cascade Classifier
Mohammad Waseem
@author: naray
"""

import cv2
import matplotlib.pyplot as plt

image_path = "face_image.png"
original_image = cv2.imread('face_image.png')

# Convert the image to grayscale for easier computation
image_grey = cv2.cvtColor(original_image, cv2.COLOR_RGB2GRAY)
print('original image')
plt.imshow(image_grey, cmap='gray')
plt.show()

# Using the Haar cascade classifier for classification.
# Internally uses adaboost.
eye_classifier = cv2.CascadeClassifier(
    f"{cv2.data.haarcascades}haarcascade_eye.xml")

face_classifier = cv2.CascadeClassifier(
    f"{cv2.data.haarcascades}haarcascade_frontalface_alt.xml")

smile_classifier = cv2.CascadeClassifier(
    f"{cv2.data.haarcascades}haarcascade_smile.xml")


detected_eyes = eye_classifier.detectMultiScale(image_grey, minSize=(50, 50))
detected_face = face_classifier.detectMultiScale(image_grey, minSize=(50, 50))
detected_smile = smile_classifier.detectMultiScale(image_grey, minSize=(200, 200))

# Draw rectangles on eyes
if len(detected_eyes) != 0:
    for (x, y, width, height) in detected_eyes:
        cv2.rectangle(original_image, (x, y),
                      (x + height, y + width),
                      (0, 255, 0), 2)
# Draw rectangles on face
if len(detected_face) != 0:
    for (x, y, width, height) in detected_face:
        cv2.rectangle(original_image, (x, y),
                      (x + height, y + width),
                      (255, 0, 0), 2)

print("Detected smile : ", detected_smile)
# Draw rectangles on smile
if len(detected_smile) != 0:
    for (x, y, width, height) in detected_smile:
        cv2.rectangle(original_image, (x, y),
                      (x + height, y + width),
                      (0, 0, 255), 2)

plt.imshow(original_image, cmap='gray')
plt.show()

# Changes for getting detected
detected_smile = smile_classifier.detectMultiScale(image_grey, minSize=(43, 100))
image_copy = original_image.copy()

print("Detected smile : ", detected_smile)
# Draw rectangles on smile
if len(detected_smile) != 0:
    for (x, y, width, height) in detected_smile:
        cv2.rectangle(image_copy, (x, y),
                      (x + height, y + width),
                      (0, 0, 255), 2)

print("With smile detection (minSize of smile classifier has been modified)")
plt.imshow(image_copy, cmap='gray')
plt.show()
```
